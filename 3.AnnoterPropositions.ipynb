```python
import random
from collections import defaultdict
import json
import pandas as pd
```

```python
#créé une liste des propositions
data = pd.read_csv('handi_newrules2.csv', encoding='utf-8')
props=list(data['content'])
```

```python
#annoter les propositions et enregistre le tout dans un fichier texte
print("nom du fichier où enregistrer les annotations ? (format json)")
filename=input()
random.shuffle(props)
prop2annots=defaultdict(list)
listedict=[]
print("nombre de propositions que l'on souhaite annoter ?")
nb=int(input())
for i in range(nb) :
    d=defaultdict(list)
    print(props[i])
    prop=props[i]
    annot=input()
    d["proposition"].append(prop)
    if annot!='s' :
        prop2annots[prop].append(annot.lower())
        d["keywords"].append(annot.lower())
    #entrez 's' quand on a fini d'annoter, pour passer à la proposition suivante
    while(annot!='s') :
        annot=input()
        if annot!='s' :
            prop2annots[prop].append(annot.lower())
            d["keywords"].append(annot.lower())
    listedict.append(d)
#liste de dicos
json_file=open(filename+'.json', 'w',encoding='utf-8')
json_file.write('[' +
        ',\n'.join(json.dumps(i) for i in listedict) +
        ']\n')
json_file.close()
```

```python
json_file2 = open(filename+'.txt',encoding='utf-8')
jsontxt = json_file2.read()
json_data = json.loads(jsontxt)
#affiche le dictionnaire des annotations, proposition --> liste de mots clés
print(json_data)
json_file.close()
json_file2.close()
```
