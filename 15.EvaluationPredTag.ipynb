```python
def rappelprecision(seuil,liste) :
    TP=0
    totalgold=0
    totalsup=0
    for elt in liste : 
        if elt[3]=='y' :
            totalgold+=1
        if elt[2]>seuil :
            totalsup+=1
        if elt[3]=='y' and elt[2]>seuil :
            TP+=1
    if totalsup==0 : 
        precision=0
    else :
        precision=TP/totalsup
    rappel=TP/totalgold
    return (precision,rappel,seuil)
```

```python
#sur autres et pas autres, à fusionner
def precision(seuil,liste) :
    TP=0
    totalsup=0
    for elt in liste : 
        if elt[2]>seuil :
            totalsup+=1
        if elt[3]=='y' and elt[2]>seuil :
            TP+=1
    if totalsup==0 : 
        precision=0
    else :
        precision=TP/totalsup
    return (precision,seuil)
```

```python
#que sur pas autres
def rappel(seuil,liste) :
    TP=0
    totalgold=0
    for elt in liste : 
        if elt[3]=='y' :
            totalgold+=1
        if elt[3]=='y' and elt[2]>seuil :
            TP+=1
    rappel=TP/totalgold
    return (rappel,seuil)
```
